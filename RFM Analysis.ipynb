#  RFM Analysis
### Recency 1 to 4 , Frequency 1 to 4 , Monetary 1 to 4

```python
import pandas as pd
import datetime as dt

new_d=pd.read_excel('Market_Basket_Data.xlsx').drop(columns=['Itemname','Country'])

today_date=dt.datetime(year=2023,month=10,day=25)

new_d['Total']=new_d['Quantity'] * new_d['Price']

rfmn=new_d.groupby('CustomerID').agg({
    'Date': lambda date: (today_date - date.max()).days,
    'BillNo': 'count',
    'Total': 'sum'
})

rfmn.columns=['Recency','Frequency','Monetary']
rfmn=rfmn[(rfmn['Monetary'] > 0 ) & (rfmn['Frequency'] > 0)]
rfmn.head()

rfmn['R_score']=pd.qcut(rfmn['Recency'],4,labels=[1,2,3,4])
rfmn['F_score']=pd.qcut(rfmn['Frequency'],4,labels=[4,3,2,1])
rfmn['M_score']=pd.qcut(rfmn['Monetary'],4,labels=[4,3,2,1])


rfmn['RFM_Scores']=(rfmn['R_score'].astype(str) +
                   rfmn['F_score'].astype(str) +
                   rfmn['M_score'].astype(str))

rfmn.info()

rfmn
# rfmn.groupby('CustomerID').agg({
# 'Recency': ['mean','min','max','count'],
# 'Frequency': ['mean','min','max','count'],
# 'Monetary': ['mean','min','max','count'] })
```

```python
# 1) Best Customers: 111 Ranking

rfmn[rfmn['RFM_Scores']=='111']
```
